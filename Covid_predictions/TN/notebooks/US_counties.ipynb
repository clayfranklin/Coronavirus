```python
# viral_times.replace({"state": abbrev_us_state})
viral_times['state']=viral_times['state'].map(us_state_abbrev)
viral_times
```

```python
viral_times['unique county']=viral_times['state']+viral_times['county']
viral_times
```

```python
TNY='https://en.wikipedia.org/wiki/User:Michael_J/County_table'
stuff=pd.read_html(TNY)
US_land_data=stuff[0]
US=US_land_data.rename(columns={'County [2]': 'County'})
US["unique county"]=US["State"]+US["County"]
US
```

```python
All_counties = US.merge(viral_times,left_on='unique county',right_on='unique county', how='outer')
All_counties.sort_values("cases")
All_counties.sort_values(["cases", "Population(2010)"], ascending=False)
lat=str(All_counties.iloc[:,12])
long=str(All_counties.iloc[:,13])
lats=[]
longs=[]
for i in lat:
    x=i[:-1]
    lats.append(x)

for steve in long:
    y=steve[:-1]
    longs.append(y)

All_counties['Latitude']=lats
All_counties["Longitude"]=longs
All_counties.reset_index(drop=True, inplace=True)
All_counties.drop(["Unnamed: 0", "Sort [1]"],axis=1)

```
