```python
import pandas as pd
```

```python
TN=pd.read_excel("C:/Users/clayf/Downloads/Public-Dataset-County-New.xlsx")
TN
```

```python
TNpop = pd.read_csv("Population Estimates by County.csv")
TNpop= TNpop.drop([96])
TNpop = TNpop.drop([0])
TNpop["Population Estimates by County"]= TNpop["Population Estimates by County"].str.split(" County", n = 1, expand = True)
TNpop = TNpop.rename(columns={'Population Estimates by County': 'County'})
TN_county_populations = TNpop.rename(columns={'Unnamed: 1': 'Population'})
TN_county_populations["Population"] = TN_county_populations["Population"].str.replace(",","").astype(int)
TN_county_populations
```

```python
TN_data = TN_county_populations.merge(TN,left_on='County',right_on='COUNTY', how='left')
TN_data.head(60)
```

```python
TN_data["Percentage of County Population"] = round((TN_data["NEW_ACTIVE"]/TN_data["Population"])*100,3)
TN_data["Percent_Tested"] = round((TN_data["TOTAL_TESTS"]/TN_data["Population"])*100,1)
TN_data
```

```python
TN_data.to_csv("C:/Users/clayf/Documents/Coronavirus/Covid_predictions/TN/Resources/Merged_TN_DATA.csv", index=False, encoding ="UTF-8")
```
